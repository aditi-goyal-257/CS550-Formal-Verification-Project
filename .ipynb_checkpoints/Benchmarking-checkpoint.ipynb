```python
import pandas as pd
```

```python
runtimes = {}
```

```python
with open("runtimes.log") as f:
    lines = f.readlines()
    for l in lines:
        a = l.split(" ")
        n = a[2]
        if not n in runtimes.keys():
            runtimes[n] = []
        timeTaken = a[5]
        runtimes[n] += [timeTaken]
    
```

```python
with open("runtimes.pkl", "rb") as f:
    r = pickle.load(f)
```

```python
df = pd.DataFrame(data = d, index = list(r.keys()))
```

```python
df['means'] = range(len(list(r.keys())))
```

```python
minObsLen = 200

for key in r.keys():
    minObsLen = min(minObsLen, len(r[key]))

for i in range(minObsLen):
    
    obs = []
    
    for key in r.keys():
        obs += [ float(runtimes[key][i]) ]
    
    df[f"obs{i}"] = obs
        
        
    
```

```python
obsColumns = [ f"obs{i}" for i in range(minObsLen) ]
```
